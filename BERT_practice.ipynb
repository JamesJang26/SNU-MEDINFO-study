```python
import pandas as pd

data_url = "https://raw.githubusercontent.com/socd06/medical-nlp/master/data/X.csv"
df = pd.read_csv(data_url)
```

```python
df.head()
```

```python
df['label'].value_counts()
```

```python
df12 = df[df['label'].isin((1,2))]
df12
```

```python
df10 = df12.copy()
df10.loc[df10['label']==2, 'label'] = 0
df10
```

```python
df10_shuffled = df10.sample(frac=1, random_state=0)
df10_shuffled
```

```python
!pip install -U datasets transformers
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df10_shuffled, preserve_index = True)

dataset = dataset.train_test_split(test_size = 0.1, seed = 0)

dataset
```

```python
from pprint import pprint
pprint(dataset['test'][42])
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

#BERT-BASE
model_path = "bert-base-uncased"

#BERT 토크나이저 가져오기
tokenizer = AutoTokenizer.from_pretrained(model_path)

#BERT 모델 가져오기
model = AutoModelForSequenceClassification.from_pretrained(model_path)
```

```python
def tokenize_many(examples):
  return tokenizer(
      examples['text'],
      max_length = model.config.max_position_embeddings,
      padding = 'max_length',
      truncation = True
  )
```

```python
dataset = dataset.map(tokenize_many, batched = True)
```

```python
import numpy as np
toy_labels = np.array([1]*5 + [0]*5)
print(toy_labels)
```

```python
np.random.seed(0)
toy_logits = np.random.randn(10, 2)
print(toy_logits)
```

```python
toy_preds = np.argmax(toy_logits, axis = -1)
print(toy_preds)
```

```python
toy_result = pd.DataFrame(zip(toy_labels, toy_preds), columns = ('True', 'Pred'))
toy_result
```

```python
def evaluate(pair):
  true, pred = pair
  head = 'T' if true == pred else 'F'
  tail = 'P' if pred else 'N'
  return head + tail

toy_result['Class'] = toy_result[['True', 'Pred']].apply(evaluate, axis = 1)
toy_result
```

```python
counts = toy_result['Class'].value_counts()
print(counts)
```
